```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/Estadisticas Sponsors.xlsx")
df = df.iloc[:, :-2]
df = df.drop(range(24, 30))
df['fuente '] = df['fuente '].str.capitalize()
df
```

#PLOTS

```python
top_sponsors = df.groupby('Nombre sponsor')['cantidad de donacion'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_sponsors.values, y=top_sponsors.index, orient='h', ci=None)
plt.xlabel('Cantidad de Donaciones')
plt.ylabel('Sponsor')
plt.title('Top 10 Sponsors con Mayor Cantidad de Donaciones')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='tipo de donacion', y='cantidad de donacion', hue='fuente ', ci=None)
plt.xlabel('Tipo de Donación')
plt.ylabel('Cantidad de Donaciones')
plt.title('Cantidad de Donaciones por Tipo y Origen')
plt.legend(title='Origen')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='conseguidos por', hue='tipo de donacion')
plt.xlabel('Conseguidor')
plt.ylabel('Cantidad de Donaciones')
plt.title('Donaciones por Tipo de Conseguidor')
plt.legend(title='Tipo de Donacion')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='conseguidos por', y='cantidad de donacion', hue='tipo de donacion', ci=None, estimator=sum)
plt.xlabel('Conseguidor')
plt.ylabel('Valor de Donaciones')
plt.title('Valor de Donaciones por Tipo de Conseguidor')
plt.legend(title='Tipo de Donacion')
plt.show()
```
